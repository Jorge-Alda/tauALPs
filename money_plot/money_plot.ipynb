```python
import matplotlib.pyplot as plt
import pandas as pd
plt.rcParams['text.usetex'] = True
```

```python
g2data = pd.read_csv('g2tau.dat', sep='\t')
Bdata = pd.read_csv('Bdecay.dat', sep='\t')
direct_data = pd.read_csv('direct_Belle.dat', sep='\t')
direct_low_data = pd.read_csv('direct_BelleII_low.dat', sep='\t')
direct_high_data = pd.read_csv('direct_BelleII_high.dat', sep='\t')
taudata = pd.read_csv('taudecay_FCCee.csv', names=['ma_GeV', 'xi'])
direct_FCCee_data = pd.read_csv('direct_FCCee.csv', names=['ma_GeV', 'xi'])
```

```python
fa = 1000
vev = 246
```

```python
plt.figure(figsize=(8, 6))
plt.plot(g2data['ma_GeV'], g2data['ctau_ATLAS_pair'], label=r"$(g-2)_\tau$ (ATLAS, $q\bar{q}\to \tau^+\tau^-$)", lw=2.5)
plt.annotate(r"$(g-2)_\tau$ (ATLAS)", (1.5e-3, 240), fontsize=14, c='tab:blue')
plt.fill_between(g2data['ma_GeV'], g2data['ctau_ATLAS_pair'], 1e7, alpha=0.2)
plt.plot(g2data['ma_GeV'], g2data['ctau_BelleII_proj'], label=r"$(g-2)_\tau$ (Belle-II)", ls='dashed', c='r')
plt.annotate(r"$(g-2)_\tau$ (Belle-II)", (1.5e-3, 4.2), fontsize=14, c='r')
plt.plot(Bdata['ma_GeV'], Bdata['ctau_Bdecay'], label=r'$B^-\to\tau^-\bar{\nu}_\tau a$', lw=2.5)
plt.annotate(r'$B^-\to\tau^-\bar{\nu}_\tau a$', (1.5e-3,1.4e3), fontsize=14, c='tab:orange')
plt.fill_between(Bdata['ma_GeV'], Bdata['ctau_Bdecay'], 1e7, alpha=0.2)
plt.plot(direct_data['ma_GeV'], direct_data['xitau_direct_Belle']*fa/vev, label=r'$e^+e^-\to\tau^+\tau^-a(a\to\ell^+\ell^-)$ (Belle)', lw=2.5)
plt.annotate(r'$e^+e^-\to\tau^+\tau^-a$' + '\n(Belle)', (0.8e-1, 10), fontsize=14, c='tab:green')
plt.fill_between(direct_data['ma_GeV'], direct_data['xitau_direct_Belle']*fa/vev, 1e7, alpha=0.2)
plt.plot(direct_low_data['ma_GeV'], direct_low_data['xitau_BelleII']*fa/vev, label=r'$e^+e^-\to\tau^+\tau^-a(a\to\ell^+\ell^-)$(Belle-II)', c='purple', ls='dashed')
plt.annotate(r'$e^+e^-\to\tau^+\tau^-a$' + '\n(Belle-II)', (2e-2, 0.3), fontsize=14, c='purple')
plt.plot(direct_high_data['ma_GeV'], direct_high_data['xitau_BelleII']*fa/vev, c='purple', ls='dashed')
plt.plot(taudata['ma_GeV'], taudata['xi']*fa/vev, label=r'$\tau^-\to\nu_\tau e^-\bar{\nu}_e$ (FCC-ee)', ls='dashed', c='k')
plt.annotate(r'$\tau^-\to\nu_\tau e^-\bar{\nu}_ea$' + '\n(FCC-ee)', (1.3, 18), fontsize=14)
plt.plot(direct_FCCee_data['ma_GeV'], direct_FCCee_data['xi']*fa/vev, label=r'$e^+e^-\to\tau^+\tau^-a(a\to\ell^+\ell^-)$ (FCC-ee)', ls='dashed', c='lightseagreen')
plt.annotate(r'$e^+e^-\to\tau^+\tau^-a$' + '\n(FCC-ee)', (2.8, 0.5), fontsize=14, c='lightseagreen')
ax = plt.gca()
ax.set_xscale('log')
ax.set_yscale('log')
plt.xlim(1e-3, 20)
plt.xlabel(r'$m_a$ [GeV]', fontsize=18)
plt.ylabel(r'$|c_{\tau\tau}|$', fontsize=18)
#plt.legend(fontsize=16)
plt.ylim(1e-1, 1e5)
plt.savefig('moneyplot.png')
```
